```python
# Import 
from tkinter import *
from tkinter import ttk
import random


root = Tk()

# geometry
root.geometry("1000x500")

# title
root.title("Rock Paper Scissor Game")

#color
root.configure(bg='pink')

# Value
computer_no = {
    "0":"Rock",
    "1":"Paper",
    "2":"Scissor"
}

#logic for game:

#1.Paper and Scissor =Scissor win

#2.Rock and Scissor =Rock win

#3.Paper and Rock =Paper win




# Restart  Game
# Disable Button
# close

def button_restart():
    btn1["state"] = "active"
    btn2["state"] = "active"
    btn3["state"] = "active"
    lbv1.config(text = "Player")
    lbv3.config(text = "Computer")
    lbv4.config(text = "")
    root.config(bg = "pink")
       

def button_disable():
    btn1["state"] = "disable"
    btn2["state"] = "disable"
    btn3["state"] = "disable"

    
    
def Close():
    root.destroy()
 
# function for :
# 1. rock
 
# 2. paper
 
# 3. scissor
def scissor():
    comp = computer_no[str(random.randint(0,2))]
    if comp == "Rock":
        game_result = " Computer Win"
    elif comp == "Scissor":
        game_result = "Match Draw"
    else:
        game_result = "you win "
    lbv4.config(text = game_result)
    lbv1.config(text = "Scissor")
    lbv3.config(text = comp)
    button_disable() 
    root.config(bg = "blue")
       

def rock():
    comp = computer_no[str(random.randint(0,2))]
    if comp == "Rock":
        game_result = "Match Draw"
    elif comp=="Scissor":
        game_result = "  you win   "
    else:
        game_result = " Computer Win"
    lbv4.config(text = game_result)
    lbv1.config(text = "Rock")
    lbv3.config(text = comp)
    button_disable()
    root.config(bg = "red")

def paper():
    comp= computer_no[str(random.randint(0, 2))]
    if comp == "Paper":
        game_result = "Match Draw"
    elif comp=="Scissor":
         game_result = " Computer Win"
    else:
        game_result = "you win"
    lbv4.config(text = game_result)
    lbv1.config(text = "Paper")
    lbv3.config(text = comp)
    button_disable()
    root.config(bg = "violetred")


#  Labels and Frames 
Label(root,text = "Saurav Gaming:Rock Paper Scissor", font = "Helvetica 22 bold",bg = "yellow", fg = "blue").pack(pady = 42)

frame = Frame(root)
frame.pack()

lbv1 = Label(frame,
      text = "Player ", font = 11,bg = "yellow")

lbv2 = Label(frame,
        text = "----------------VS--------------- ", font = "normal 12 bold",bg = "yellow")

lbv3 = Label(frame, text = "Computer", font = 11,bg = "yellow")

lbv1.pack(side = LEFT)
lbv2.pack(side = LEFT)
lbv3.pack()

lbv4 = Label(root,text = "",font = "normal 20 bold", width = 15 ,borderwidth = 2,bg = "white",    relief = "solid")
lbv4.pack(pady = 22)




#Define the button for rock,
#paper, scissor.



new_frame = Frame(root)
new_frame.pack()

btn1 = Button(new_frame, text = "Rock",font = 12, width = 8,bg = "yellow",command = rock)

btn2 = Button(new_frame, text = "Paper ",font = 12, width = 8,bg = "yellow",command = paper)

btn3 = Button(new_frame, text = "Scissor", font = 12, width = 8,bg = "yellow", command = scissor)

btn1.pack(side = LEFT, padx = 15)
btn2.pack(side = LEFT,padx = 15)
btn3.pack(padx = 15)

Button(root, text = "Restart",font = 12, fg = "red",bg = "yellow", command =button_restart ).pack(pady = 25)
Button(root, text = "Exit",font = 12, fg = "red",bg = "yellow", command = Close).pack(pady = 25)


root.mainloop()
```
